# Standard prelude

```python
import numpy as np
import scipy.linalg as la
from matplotlib import pyplot as plt
import scipy.special

%matplotlib inline
%config InlineBackend.figure_format='retina'
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.rcParams['image.interpolation'] = 'nearest'
```

# Interpolation using Legendre polynomials

```python
years = np.linspace(1979, 2014, 36)
years_fine = np.linspace(1979, 2014, 300)
population = np.array([128808, 128571, 128525, 128165, 128199, 127952, 127128, 126050, 126475, 126976, 
                       126699, 127515, 127118, 126058, 125457, 125264, 124561, 123577, 123325, 124205, 
                       125174, 124835, 125485, 126441, 126825, 127194, 127597, 128228, 129268, 130721, 
                       133265, 134753, 136288, 137586, 139390, 140228], dtype=np.float64)
plt.plot(years, population, 'o');
```

```python
degree = 10

A = np.column_stack([years ** i for i in range(degree)])
np.linalg.cond(A)
```

```python
pos = np.linspace(-1, 1, 36)
A = np.column_stack([pos ** i for i in range(degree)])
np.linalg.cond(A)
```

```python
A = np.column_stack([scipy.special.legendre(i)(pos) for i in range(degree)])
np.linalg.cond(A)
```

```python
Q, R = la.qr(A, mode='economic')
print(Q.shape)
print(R.shape)
x = la.solve_triangular(R, Q.T @ population)


pos_fine = np.linspace(-1, 1, 100);
A_fine = np.column_stack([scipy.special.legendre(i)(pos_fine) for i in range(degree)])
plt.plot(A_fine)
result = A_fine @ x
```

```python
plt.plot(years, population, 'o');
plt.plot((pos_fine + 1) * 35/2 + 1979, result)
```

# Spline interpolation

```python
from scipy.interpolate import interp1d
```

```python
f = interp1d(years, population, kind=3)
```

```python
plt.plot(years, population, 'o');
plt.plot(years_fine, f(years_fine));
```

# Radial basis function interpolation

```python
from scipy.interpolate import Rbf
plt.rcParams['figure.figsize'] = (8.0, 8.0)
```

```python
def f(x, y):
    return x*np.exp(-x**2-y**2)
ti = np.linspace(-2.0, 2.0, 100)
xg, yg = np.meshgrid(ti, ti)
plt.pcolor(xg, yg, f(xg, yg))
```

```python
n = 40
x = np.random.rand(n)*4.0-2.0
y = np.random.rand(n)*4.0-2.0
plt.scatter(x, y, c='k')
```

```python
interp = Rbf(x, y, f(x, y), epsilon=2)
plt.pcolor(xg, yg, interp(xg, yg))
plt.scatter(x, y, c='k')
```

```python
plt.pcolor(xg, yg, (interp(xg, yg)-f(xg, yg)), cmap='coolwarm',vmin=-0.1, vmax=0.1)
plt.colorbar()
```
